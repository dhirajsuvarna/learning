```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
print(plt.style.available)
#plt.style.use('seaborn-darkgrid')
plt.xkcd()
```

## Stateful Approach of Matplotlib

```python
age = df['Age']
all_devs = df['All_Devs']
py_devs = df['Python']
js_devs = df['JavaScript']

plt.plot(age, all_devs, label="All Devs", linestyle='--', color='r')
plt.plot(age, py_devs, label='Python')
plt.plot(age, js_devs, label='JavaScript')

plt.legend()

plt.xlabel('Age')
plt.ylabel('Salary (USD)')

plt.title('Salary of Devs')
plt.grid()
plt.tight_layout()
plt.savefig('plot.png')
```

## Object Oriented Approch of Matplotlib

```python
fig, ax = plt.subplots() #--> Very Important

ax.plot(age, all_devs, label="All Devs", linestyle='--', color='r')
ax.plot(age, py_devs, label='Python')
ax.plot(age, js_devs, label='JavaScript')

ax.set_xlabel('Umar')
ax.set_ylabel('Aukad (USD)')
ax.legend()

plt.title('Salary of Devs')
plt.grid()
plt.tight_layout()
#plt.show()
```

### Let's create a different plot of each tech

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=1, nrows=3)

ax1.plot(age, all_devs, label="All Devs", linestyle='--', color='r')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()

ax2.plot(age, py_devs, label='Python Dev')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary')
ax2.legend()

ax3.plot(age, js_devs, label='Javascript Dev')
ax3.set_xlabel('Age')
ax3.set_ylabel('Salary')
ax3.legend()

ax1.set_title('yo graph')
plt.tight_layout()
```
