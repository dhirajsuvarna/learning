## Exercise 10: Merge two data frames using the following condition

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
carprice_df = pd.DataFrame(Car_Price)
carhp_df = pd.DataFrame(car_Horsepower)
```

```python
carprice_df
```

```python
carhp_df
```

```python
pd.concat([carprice_df, carhp_df], axis=1)
```

```python
cars_df = pd.merge(carprice_df, carhp_df, on="Company")
cars_df
```
